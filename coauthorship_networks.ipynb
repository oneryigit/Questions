```
import numpy
import pandas
import networkx as nx
import matplotlib.pyplot as plt
```

I am working on coauthorship network using data from this repository: [link](http://www-personal.umich.edu/~mejn/netdata/).

**Node** represents authors.

**Edge** represents coauthorship.

```
# I am using gml data type.

network_file = '/content/netscience.gml'
H = nx.read_gml(network_file)
```

```
print(nx.number_of_nodes(H)) #node counts
```

```
print(nx.number_of_edges(H)) #edge counts
```

The network is undirected.

```
G = nx.Graph(H)


fig, ax= plt.subplots(figsize=(10,10))
nx.draw_networkx(G, ax=ax, font_size=4, node_size=20,
                 edge_color='r',width=2, alpha=0.8)
plt.xlim(-0.7, 0.7)
plt.ylim(-0.7, 0.7)
plt.axis("off")  # turn off axis
plt.show()
```
